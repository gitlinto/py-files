**Pandas Series**

```python
import pandas as pd
import numpy as np

# Option 1: Create a Series from a List
data_list = [10, 20, 30, 40]
series_from_list = pd.Series(data_list)
print("Series from a List:")
print(series_from_list)
print()

# Option 2: Create a Series with a Custom Index
data_custom_index = [100, 200, 300, 400]
index_custom = ['a', 'b', 'c', 'd']
series_with_custom_index = pd.Series(data_custom_index, index=index_custom)
print("Series with Custom Index:")
print(series_with_custom_index)
print()

# Option 3: Create a Series from a Dictionary
data_dict = {'Alice': 25, 'Bob': 30, 'Charlie': 22}
series_from_dict = pd.Series(data_dict)
print("Series from a Dictionary:")
print(series_from_dict)
print()

# Option 4: Create a Series with a Scalar Value
scalar_value = 5
series_from_scalar = pd.Series(scalar_value, index=['a', 'b', 'c', 'd'])
print("Series from a Scalar Value:")
print(series_from_scalar)
print()

# Option 5: Create a Series from a NumPy Array
data_numpy_array = np.array([1.1, 2.2, 3.3, 4.4])
series_from_numpy_array = pd.Series(data_numpy_array)
print("Series from a NumPy Array:")
print(series_from_numpy_array)
print()

# Option 6: Create a Series from a List of Strings
data_strings = ['apple', 'banana', 'cherry']
series_from_strings = pd.Series(data_strings)
print("Series from a List of Strings:")
print(series_from_strings)
print()

# Option 7: Create a Series with Missing Data (NaN)
data_with_nan = [10, None, 30, np.nan]
series_with_nan = pd.Series(data_with_nan)
print("Series with Missing Data (NaN):")
print(series_with_nan)
print()
```

1. Indexing:
**Series**: Each element is associated with an explicit index (which can be customized, not necessarily integers).
**Array**: Elements are indexed by integers only, starting from 0.

2. Data Labeling:
**Series**: Supports labeled data, so each element can have a corresponding label (index).
**Array**: Contains raw data without labels; the elements are accessed by their position.

3. Data Type:
**Series**: Can hold mixed data types (e.g., numbers, strings, etc.) within a single Series.
**Array**: Usually contains elements of the same data type, making it more efficient for numerical computations.

4. Alignment:
**Series**: When performing operations, pandas automatically aligns elements based on their index labels, allowing for element-wise operations on labeled data.
**Array**: Operations are based purely on the position of the elements, without considering labels.

5. Metadata:
**Series**: Stores additional metadata like the index.
**Array**: No extra metadata, just raw values.

6. Functionality:
**Series:** Built on top of NumPy arrays but includes powerful features like handling missing data (NaN), alignment on indices, and more flexibility for data manipulation.

**Array:** NumPy arrays are more optimized for performance, especially for mathematical computations, but lack the flexibility of labels and missing value handling that Series offers.

```python
import numpy as np

array = np.array([10, 20, 30, 40])
print(array[1])  # Output: 20 (accesses the second element by its index)
```

```python
import pandas as pd

series = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(series[1])      # Output: 20 (position-based)
print(series['b'])    # Output: 20 (label-based)
```

**LOC AND ILOC**

In pandas, loc and iloc are two powerful methods used for selecting data from a DataFrame. They differ in how they index and access the data:

1. loc[]: Label-based indexing
Description: loc is used to access **rows and columns** by their labels (names). It allows you to reference rows and columns using the actual index labels and column names.

2. Slicing behavior: When slicing with loc, both the start and end labels are inclusive.

Usage example:
python

Select row with label 'AA' and columns 'Name' and 'Age'
df.loc['AA', ['Name', 'Age']]

3. When to use: Use loc when you want to select data by specific row/column labels, or when working with a DataFrame where the index (row labels) has meaningful names.

iloc[]: Integer-based indexing

1. Description: iloc is used to access rows and columns by their integer positions. It does not care about the actual labels but instead uses the zero-based integer positions of rows and columns.

2. Slicing behavior: When slicing with iloc, the end index is exclusive, similar to Python's standard slicing behavior.
Usage example:
python

Select the first row (index position 0) and first two columns (index positions 0 and 1)
df.iloc[0, [0, 1]]

3. When to use: Use iloc when you need to access data based on its position in the DataFrame, especially when working with numerical indices or when the order of rows/columns is more important than their labels.

**Joins in Pandas**

**Inner Join**

```python
import pandas as pd

# create the first DataFrame
df1 = pd.DataFrame({
    'employee_id': ['1', '2', '3', '4'],
    'name': ['John', 'Mary', 'David', 'Sarah']
})

# create the second DataFrame
df2 = pd.DataFrame({
    'employee_id': ['2', '3', '4', '5'],
    'department': ['HR', 'Finance', 'IT', 'Marketing']
})

# perform an inner join on the 'employee_id' column
result = pd.merge(df1, df2, on='employee_id', how='inner')

print(result)
```

**Left Join**

```python
import pandas as pd

# create the first DataFrame
df1 = pd.DataFrame({
    'employee_id': ['1', '2', '3', '4'],
    'name': ['John', 'Mary', 'David', 'Sarah']
})

# create the second DataFrame
df2 = pd.DataFrame({
    'employee_id': ['2', '3', '4', '5'],
    'department': ['HR', 'Finance', 'IT', 'Marketing']
})

# perform a left join on the 'employee_id' column
result = pd.merge(df1, df2, on='employee_id', how='left')

print(result)
```

**Right Join**

```python
import pandas as pd

# create the first DataFrame
df1 = pd.DataFrame({
    'employee_id': ['1', '2', '3', '4'],
    'name': ['John', 'Mary', 'David', 'Sarah']
})

# create the second DataFrame
df2 = pd.DataFrame({
    'employee_id': ['2', '3', '4', '5'],
    'department': ['HR', 'Finance', 'IT', 'Marketing']
})

# perform a right join on the 'employee_id' column
result = pd.merge(df1, df2, on='employee_id', how='right')

print(result)
```

**Outer Join**

```python
import pandas as pd

# create the first DataFrame
df1 = pd.DataFrame({
    'employee_id': ['1', '2', '3', '4'],
    'name': ['John', 'Mary', 'David', 'Sarah']
})

# create the second DataFrame
df2 = pd.DataFrame({
    'employee_id': ['2', '3', '4', '5'],
    'department': ['HR', 'Finance', 'IT', 'Marketing']
})

# perform an outer join on the 'employee_id' column
result = pd.merge(df1, df2, on='employee_id', how='outer')

print(result)
```

**Groupby Function**

The groupby function in Pandas is used to split data into groups based on certain criteria, apply a function to each group independently, and then combine the results.

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Edward', 'Fiona', 'George', 'Helen', 'Ian', 'Jack'],
    'Department': ['HR', 'Finance', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'HR', 'IT', 'Finance'],
    'Salary': [50000, 60000, 55000, 65000, 70000, 62000, 48000, 57000, 52000, 64000],
    'Age': [25, 35, 45, 32, 28, 38, 40, 29, 36, 27],
    'Gender': ['F', 'M', 'M', 'M', 'M', 'F', 'M', 'F', 'M', 'M']
}

df = pd.DataFrame(data)
df
```

**Example 1: Group By a Single Column and Calculate Summary Statistics**


Let's group by the "Department" column and calculate the average salary in each department.

df.groupby('column_name').aggregation_function()


```python
df.groupby('Department')['Salary'].mean()
```

```python
# Average salary by department
avg_salary_by_department = df.groupby('Department')['Salary'].mean()
print(avg_salary_by_department)
```

**Example 2: Group By Multiple Columns**

```python
# Average age by department and gender
avg_age_by_dept_gender = df.groupby(['Department', 'Gender'])['Age'].mean()
print(avg_age_by_dept_gender)
```

In Pandas, you can group by multiple columns by passing a list of column names to the groupby() function. When you group by multiple columns, Pandas creates a hierarchical index, where each unique combination of the values in these columns forms a separate group

**Example 3: Aggregating Multiple Functions on Different Columns**

You can use .agg() to apply multiple aggregation functions to multiple columns. For example, lets calculate both the mean and the maximum salary and age by Department.

```python
# Mean and maximum salary and age by department
salary_age_stats = df.groupby('Department').agg({'Salary': ['mean', 'max'], 'Age': ['mean', 'max']})
print(salary_age_stats)
```

**Example 4: Filtering Groups Based on a Condition**

Lets say we want to find departments where the average salary is above 59000.

```python
# Filter departments with an average salary above 59000
high_salary_departments = df.groupby('Department').filter(lambda x: x['Salary'].mean() > 59000)
print(high_salary_departments)
```

lambda arguments : expression

filter(function, iterable)

**Example 5: Transforming Data Within Groups**

Suppose you want to add a column that shows each persons salary as a percentage of the average salary in their department.

```python
# Salary as a percentage of the department's average salary
df['Salary_Percent_of_Dept_Avg'] = df.groupby('Department')['Salary'].transform(lambda x: x / x.mean() * 100)
print(df[['Name', 'Department', 'Salary', 'Salary_Percent_of_Dept_Avg']])
```

```python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [24, 27, 22, 29]
})

# Inner merge on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print(merged_df)
```

image.png

```python
# Create two DataFrames with different information but the same index
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
}, index=[1, 2, 3, 4])

df2 = pd.DataFrame({
    'Age': [24, 27, 22, 29]
}, index=[1, 2, 3, 4])

# Join df2 to df1 using index
joined_df = df1.join(df2)
print(joined_df)
```

```python
# Create two DataFrames with the same columns
df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Charlie', 'David']
})

# Concatenate row-wise (stack rows)
concat_df = pd.concat([df1, df2], axis=0)
print(concat_df)
```

```python
# Create two DataFrames with the same index
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame({
    'Age': [24, 27]
})

# Concatenate column-wise (side by side)
concat_df = pd.concat([df1, df2], axis=1)
print(concat_df)
```

**rolling**

In Pandas, rolling is used to perform rolling window calculations on time series data or any sequential data. This method allows you to compute statistics (such as mean, sum, median, etc.) over a fixed-size sliding window across the data. Its particularly useful for analyzing trends over time, smoothing data, and calculating moving averages.

DataFrame.rolling(window, min_periods=None, center=False, win_type=None, on=None, axis=0).aggregate_function()


image.png

Example of rolling with Moving Average
Lets look at a few examples to understand how rolling works.

Example 1: Calculating a Simple Moving Average
Here, we'll calculate the 3-day moving average on a time series data.

```python
import pandas as pd

# Sample time series data
data = {
    'Day': [1, 2, 3, 4, 5, 6, 7],
    'Temperature': [30, 32, 34, 33, 31, 29, 28]
}
df = pd.DataFrame(data)

# Calculate the 3-day moving average for Temperature
df['3_day_MA'] = df['Temperature'].rolling(window=3).mean()
print(df)
```

Example 2: Rolling Sum
You can also calculate a rolling sum instead of a mean.

```python
# Calculate the 3-day rolling sum for Temperature
df['3_day_Sum'] = df['Temperature'].rolling(window=3).sum()
print(df)
```

Example 3: Using min_periods Parameter

min_periods: Allows you to control the minimum number of observations required for a valid result.

which means that we only need 1 value in the window to compute a result.

```python
# Calculate the 3-day moving average with a minimum of 1 non-NaN observation
df['3_day_MA_min_1'] = df['Temperature'].rolling(window=3, min_periods=1).mean()
print(df)
```

```python
# Calculate the 3-day moving average with a minimum of 2 non-NaN observation
df['3_day_MA_min_1'] = df['Temperature'].rolling(window=3, min_periods=2).mean()
print(df)
```

Example 4: Rolling Standard Deviation

```python
# Calculate the 3-day rolling standard deviation for Temperature
df['3_day_Std'] = df['Temperature'].rolling(window=3).std()
print(df)
```
